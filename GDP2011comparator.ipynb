Now we have both tables but with different size. It means that the countries where not the same in 2010 as in 2011. To avoid that what we do is delete the rows containing the information about countries that did not appear in both tables. We know that those rows are the following in prediction dataframe: (3,5,10,36,40,50,53,72,77,83,85,104,126,149,154,158,164,185,193,198,206,215,228,241). Those values are substracted from R analysis.

```python
notcountry=[3,5,10,36,40,50,53,72,77,83,85,104,126,149,154,158,164,185,193,198,206,215,228,241]
```

```python
prediction=prediction.drop(prediction.index[notcountry])
prediction
```

Now we have to order the dataframes alphabetically to be able to rest themselves.

```python
prediction=prediction.sort_values('CountryCode')
prediction
```

```python
prediction=prediction.rename(columns={'0': 'ValuePred'}).reset_index()
```

```python
prediction=prediction.drop(columns=['index'])
```

```python
prediction=prediction.drop(columns=['level_0'])
```

```python
prediction
```

```python
real_values=real_values.sort_values('CountryCode')
real_values
```

```python
real_values=real_values.reset_index()
real_values=real_values.drop(columns=['index'])
real_values
```

```python
real_values=real_values.drop(columns=['level_0'])
real_values
```

Finally we have both dataframes with the same dimensions and same CountryCode's. We just need to rest themselves.

```python
data_frame_total=pd.DataFrame([prediction['CountryCode'],prediction['ValuePred'],real_values['Value']]).transpose()
data_frame_total
```

```python
Differences=data_frame_total["ValuePred"]-data_frame_total["Value"]
Differences
```

```python
data_frame_total.insert(3, "Rest", Differences, True) 
data_frame_total
```

So here we have the column "Rest" which is the difference between the real value and the predicted one.

```python
column=data_frame_total['Rest']
max_value=column.max()
min_value=column.min()
print(min_value)
print(max_value)
```
